```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
```

```python
Data = pd.read_csv(r'C:\Users\rohit\Desktop\RAJAT SAHU/Data.csv')
Data.drop('Unnamed: 0',axis=1,inplace=True)
Data0 = Data[Data.ChurnDuration==0]
Data3= Data[Data.ChurnDuration==3]
print(Data0.shape)
print(Data3.shape)
```

```python
print("Total number "+str(len(Data0)+len(Data3))+" of users in 0th and 3rd churn in mentioned 15 Days.")
print("Total number "+str(Data0.shape[0])+" of Total Member of 0th Churn day in mentioned 15 Days.")
print("Total number "+str(Data3.shape[0])+" of Total Member of 3rd Churn day in mentioned 15 Days.")
data=[Data0.shape[0],Data3.shape[0]]
labels=['0th Churn Ratio','3rd Churn Ratio']
colors = sns.color_palette('pastel')[0:2]
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
print("Total number "+str(len(Data0)+len(Data3))+" of users in 0th and 3rd churn in mentioned 15 Days.")
print("Total number "+str(Data0.shape[0])+" of Total Member of 0th Churn day in mentioned 15 Days.")
print("Total number "+str(Data3.shape[0])+" of Total Member of 3rd Churn day in mentioned 15 Days.")
```

# JOINING BONUS

```python
JoiningBonus = Data0[Data0['JoiningBonus'] == 1]
NotJoiningBonus = Data3[Data3['JoiningBonus'] == 1]
print("Total number "+str(len(JoiningBonus)+len(NotJoiningBonus))+" of users who got joining\nbonus in 0th and 3rd churn in mentioned 15 Days.")
print("0th churn :-",len(JoiningBonus))
print("3rd churn :-",len(NotJoiningBonus))
data=[JoiningBonus.shape[0],NotJoiningBonus.shape[0]]
labels=['0th churn','Got 3rd churn']
colors = sns.color_palette("husl")[0:3]
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
Played = Data0[(Data0['JoiningBonus'] == 1) & (Data0['TotalGamePlayed_2'] >= 1)]
NotPlayed = Data3[(Data3['JoiningBonus'] == 1) & (Data3['TotalGamePlayed_2'] >= 1)]
print("Total number "+str(len(Played)+len(NotPlayed))+" of users who got joining bonus\nand Played in 0th and 3rd churn in mentioned 15 Days.")
print("0th churn :-",len(Played))
print("3rd churn :-",len(NotPlayed))
data=[Played.shape[0],NotPlayed.shape[0]]
labels=['0th churn played','3rd churn Played']
colors = sns.color_palette("husl")[0:3]
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
#WHO USERS GETTING JOINING BONUS AND PLAYED DISTRIBUTION
DATA0 = Data0[(Data0['JoiningBonus'] == 1) & (Data0['TotalGamePlayed_2'] >= 1)]
DATA3 = Data3[(Data3['JoiningBonus'] == 1) & (Data3['TotalGamePlayed_2'] >= 1)]
temp1 = pd.DataFrame(DATA0[['CardGamePlayed','FantasyPlayed','CasualPaid','CasualFree','PlayedTournament']].sum())
temp2 = pd.DataFrame(DATA3[['CardGamePlayed','FantasyPlayed','CasualPaid','CasualFree','PlayedTournament']].sum())
final1 = temp1.reset_index()
final2 = temp2.reset_index()
Header = final1['index'].to_list()
value_played = final1[0].to_list()
df0=pd.DataFrame({'Header':Header, 'Total Users Played':value_played})
Header = final2['index'].to_list()
value_played = final2[0].to_list()
df3=pd.DataFrame({'Header':Header, 'Total Users Played':value_played}) 
x = Header
y0=df0['Total Users Played']
y3=df3['Total Users Played']
plot = go.Figure(data=[go.Bar(name = '3rd churn',x = x,y = y3),go.Bar(name = '0th churn',x = x,y = y0)])
plot.update_layout(barmode='stack')                  
plot.show()
```
