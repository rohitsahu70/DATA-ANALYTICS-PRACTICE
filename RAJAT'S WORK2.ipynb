# BEFORE LABELING

```python
list1=list(df.columns)
plt.figure(figsize=(15,50))
for i,column in enumerate(list1):
    plt.subplot(len(list1), 2, i+1)
    plt.suptitle("PLOT ", fontsize=20, x=0.5, y=1)
    (df[column].value_counts(normalize=True)*100).plot()
    plt.tight_layout()
```

```python
list1=list(df.columns)
plt.figure(figsize=(15,50))
for i,column in enumerate(list1):
    plt.subplot(len(list1), 2, i+1)
    plt.suptitle("COUNT PLOT ", fontsize=20, x=0.5, y=1)
    sns.countplot(df[column])
#     plt.title(f"{column}")
    plt.tight_layout()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),annot=True)
```

```python
# list1=['age','Nodalstatus']
list1=list(df.columns)
plt.figure(figsize=(8,35))
for i,column in enumerate(list1):
    plt.subplot(len(list1), 2, i+1)
    plt.suptitle("Column Distribution ", fontsize=20, x=0.5, y=1)
    sns.distplot(df[column])
    plt.tight_layout()
```

# AFTER LABELING

```python
list1=list(df.columns)
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
for col in list1:
    df[col]=label_encoder.fit_transform(df[col])
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
```
